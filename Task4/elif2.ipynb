How many orders were made in the period

How many items were orderd in total

```python
total_items_ordered=chipo['quantity'].sum()
total_items_ordered
```

List the 5 most-ordered items

```python
most_ordered_items = chipo.groupby('item_name')['quantity'].sum().sort_values(ascending=False).head(5)
most_ordered_items
```

List the 3 most ordered item in the choice_description column

```python
most_ordered_choices = chipo.groupby('choice_description')['quantity'].sum().sort_values(ascending=False).head(3)
most_ordered_choices
```

List the top 5 items by the revenue

```python
chipo['revenue'] = chipo['quantity'] * chipo['item_price']
top_5_revenue_items = chipo.groupby('item_name')['revenue'].sum().sort_values(ascending=False).head(5)
top_5_revenue_items
```

How much was the revenue for the period in the dataset

```python
total_revenue = chipo['revenue'].sum()
total_revenue
```

What is the average revenue amount per order

```python
print(chipo['item_price'].dtype)

chipo['item_price'] = chipo['item_price'].replace('[\$,]', '', regex=True).astype(float)
chipo['revenue'] = chipo['quantity'] * chipo['item_price']
average_revenue_per_order = chipo.groupby('order_id')['revenue'].sum().mean()
print(average_revenue_per_order)


```

Select the orders which item_price is bigger than 10 dolalrs and lesser then 15 dollars



```python
chipo['item_price'] = chipo['item_price'].replace('[\$,]', '', regex=True).astype(float)

# Boş değerleri kontrol edin ve doldurun
chipo['item_price'].fillna(0, inplace=True)

# Filtreleme işlemi
filtered_orders = chipo[(chipo['item_price'] > 10) & (chipo['item_price'] < 15)]
print(filtered_orders)
```
