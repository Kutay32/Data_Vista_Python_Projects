How many orders were made in the period

```python
#Benzersiz değerler için nuniqe() metodu kullanılır.

uniqe_orders = chipo['order_id'].nunique()
uniqe_orders
```

How many items were orderd in total

```python
total_items_ordered = chipo['quantity'].sum()
total_items_ordered
```

List the 5 most-ordered items

```python
#En çok sipariş edilen itemi bulmak için item_name ve quantity sütünlarını gruplamayıp toplamam gerekiyor
grouped = chipo.groupby('item_name')['quantity'].sum()

top_5_items = grouped.nlargest(5)
print(top_5_items)
```

List the 3 most ordered item in the choice_description column

```python
#Seri üzerindeki uniqe değerleri saymak için value_counts() metodu kullanılır
top_choices = chipo['choice_description'].value_counts()
top_choices.nlargest(3)
```

List the top 5 items by the revenue

```python
#DataType Control
print(chipo['item_price'].dtype)
print(chipo['quantity'].dtype)

chipo['revenue'] = chipo['quantity'] * chipo['item_price']

top_5_revenue = chipo.groupby('item_name')['revenue'].sum().nlargest(5)
top_5_revenue

```

How much was the revenue for the period in the dataset

```python
total_revenue = chipo['revenue'].sum()
print("Total revenue is: $", total_revenue)
```

What is the average revenue amount per order

```python
# Ortalama gelir = Toplam gelir / toplam sipariş

total_revenue = chipo['revenue'].sum()
total_orders = chipo['order_id'].nunique()

print("Total revenue is: $", total_revenue)
print("Total orders is: ", total_orders)

average_revenue_per_order = total_revenue / total_orders
print("Average revenue amount per order: $", average_revenue_per_order)
```
