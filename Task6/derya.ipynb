```
image=cv2.imread("/shocked_guy (1).png")

image_rgb=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("RGB Image")
plt.show()
```

```
flipped_image = cv2.flip(image_rgb ,0)
plt.imshow(flipped_image)
```

```
image_copy = image_rgb.copy()
start_point = (1100, 900)
end_point = (400, 0)
color = (255, 0, 0)
thickness = 5
cv2.rectangle(image_copy, start_point, end_point, color, thickness)
plt.imshow(image_copy)
```

```
image_cat= cv2.imread("/shocked_cat (1).jpg",0)
plt.imshow(image_cat,cmap='gray')
plt.show()
```

```
threshold_value, binary_image = cv2.threshold(image_cat, 100 ,130,cv2.THRESH_BINARY)
plt.imshow(binary_image, cmap='gray')
plt.show()
```
