Open the "shocked_guy" image and display it in the notebook. Make sure to correct for the RGB order. Try to read and correct RGB in same line

```python
image = cv2.imread("shocked_guy.png")

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("RGB Image")
plt.axis('off')
plt.show()
```

Flip the image upside down and display it in the notebook.

```python
flipped_img = cv2.flip(img_rgb, 0)
plt.imshow(flipped_img)
plt.axis('off')
plt.show()
```

Copy the image and draw an empty RED rectangle around the human face on coppied image and display the image in the notebook.

```python

img_copy = img_rgb.copy()
start_point = (150, 300)
end_point = (380, 30)
color = (255, 0, 0)
thickness = 3
cv2.rectangle(img_copy, start_point, end_point, color, thickness)

plt.imshow(img_copy)
plt.axis('off')
plt.title("red rectangle added")
plt.show()
```

Open the "shocked_cat" image and display it in the notebook as black and white

```python
cat_img = cv2.imread("shocked_cat.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(cat_img,cmap='gray')

plt.show()

```

Apply a binary 'gray' threshold onto the image.

```python
threshold_value, binary_img = cv2.threshold(cat_img, 100 ,130,cv2.THRESH_BINARY)
plt.imshow(binary_img, cmap='gray')
plt.show()
```

Create a Horizontal Sobel Filter with a kernel size of 5 to the grayscale version of the giaraffes image and then display the resulting gradient filtered version of the image.

```python
giraffe_img = cv2.imread("giraffes.jpg", cv2.IMREAD_GRAYSCALE)
sobel_horizontal = cv2.Sobel(giraffe_img, cv2.CV_64F, 1, 0, ksize=5)


plt.imshow(sobel_horizontal, cmap='gray')
plt.show()
```
