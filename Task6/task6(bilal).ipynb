```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread("shocked_monkey.jpg")
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
img[:,:,0]=0
img[:,:,1]=0
plt.imshow(img)
```

```python
img2=cv2.imread("shocked_guy.png")
img2_rgb=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```python
new_img2=img2_rgb.copy()
new_img2=cv2.flip(new_img2,0)
plt.imshow(new_img2)
```

```python
drawimg=new_img2.copy()
cv2.rectangle(drawimg,pt1=(400,300),pt2=(1200,1350),color=(255,0,0),thickness=10)
plt.imshow(drawimg)
```

```python
img3 = cv2.imread("shocked_cat.jpg",0)
plt.imshow(img3,cmap="gray")
```

```python
ret,thresh=cv2.threshold(img3,127,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh,cmap="gray")
```

```python
img5=cv2.imread("giraffes.jpg")
sobelimg5=cv2.Sobel(img5,cv2.CV_64F,1,0,ksize=5)
plt.imshow(sobelimg5)
```
