2) Use cv2 and matplotlib to read and display the 'monkey_image'.

```python
def show_image(image, title="Görüntü", cmap=None):
    plt.imshow(image, cmap=cmap)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

3) Correct the color type to RGB using cv2 and show the result

```python

monkey_image = cv2.imread('shocked_monkey.jpg')  


monkey_image_rgb = cv2.cvtColor(monkey_image, cv2.COLOR_BGR2RGB)

show_image(monkey_image_rgb, title="Orijinal Monkey Görüntüsü (RGB)")
```

10) Create a Horizontal Sobel Filter with a kernel size of 5 to the grayscale version of the giaraffes image and then display the resulting gradient filtered version of the image.

```python
giraffes_image = cv2.imread('giraffes.jpg', cv2.IMREAD_GRAYSCALE)
sobel_horizontal = cv2.Sobel(giraffes_image, cv2.CV_64F, 1, 0, ksize=5)
show_image(np.abs(sobel_horizontal), title="Yatay Sobel Filtresi Sonucu", cmap="gray")
```
